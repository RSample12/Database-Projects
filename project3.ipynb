#### 2. create and populate BQ tables:

```python
import os
```

```python
gsutil_cmd = "gsutil ls gs://cs327e-open-access-2/school-enrollments/*"
file_listings = os.popen(gsutil_cmd)
dataset="school_enrollments"

for file in file_listings:

    start_index = file.rindex("_") + 1
    end_index = file.rindex(".")
    table = file[start_index:end_index]
    
    if table in ("co", "district", "fl", "ga", "id", "ma", "mo", "ne", "ny", "or", "school", "sd", "wi"):
        print("skipping " + table)
        continue

    bq_cmd = "bq --location=US load --autodetect --skip_leading_rows=1 "\
             "--source_format=CSV " + dataset + "." + table + ' ' + file
    print(bq_cmd)
    
    os.system(bq_cmd)
```

### 1. Sum the number of students recieving free and reduced lunch within Los Angeles county school districts for the 2020 school year, only considering k-12 schools.

```python
%%bigquery
select district, sum(cast(free_and_reduced as float64)) as free_and_reduced
from school_enrollments.ca
where county = 'LOS ANGELES' and year = 2020 and free_and_reduced != '' and grade = 'k_12_total'
group by district
order by district
```

### 2. List all counties in Texas, counting the number of schools and summing the number of students within each county.

```python
%%bigquery
select county, count(school) as schools, sum(safe_cast(total as int64)) as students
from school_enrollments.tx
where admin_level != 'district'
group by county
order by county
```

### 3. Select the state, school district and the sum of English language learners in each school district in California and Minnesota, and join the results.

```python
%%bigquery
select state, district, year, sum(safe_cast(ell as float64)) as ell
from school_enrollments.mn
where admin_level = 'school'
group by state, district, year
UNION ALL
select state, district, year, sum(safe_cast(ell as float64)) as ell
from school_enrollments.ca
where admin_level = 'school'
group by state, district, year
order by district
```

### 4. Select the avg number of students across schools per county on free and reduced lunch ordered by avg number of students, showing only those that have >200 

```python
%%bigquery
select county, year, avg(cast(free_and_reduced as float64)) as avg_free_and_reduced
from school_enrollments.ca
where free_and_reduced != '' and admin_level = 'school'
group by county, year
having avg_free_and_reduced > 200
order by avg_free_and_reduced
```

### 5. Select the district, state, year, sum of students, sum of homeless, sum of low_income students in Washington and California, grouping by district, state and year where the number of homeless students was over 300.

```python
%%bigquery
select district, state, year, sum(safe_cast(total as int64)) as total, sum(safe_cast(homeless as int64)) as homeless, sum(safe_cast(low_income as int64)) as low_income
from school_enrollments.wa
where admin_level != 'district'
group by district, state, year
having homeless > 300
UNION ALL
select district, state, year, sum(safe_cast(total as int64)) as total, sum(safe_cast(homeless as int64)) as homeless, sum(safe_cast(low_income as int64)) as low_income
from school_enrollments.ca
group by district, state, year
having homeless > 300
order by total ASC
```

### 6. Find the number of white students and the total number for each district in Texas, only showing districts with more than 10,000 students. 

```python
%%bigquery
select sum(safe_cast(white as int64)) as white, sum(safe_cast(total as int64)) as total, district, year
from school_enrollments.tx
where admin_level != 'district'
group by district, year
having total > 10000
order by total
```

### 7. Find the number of English learners in school districts and grades in California, only showing those which had more than 1000 ELL students.

```python
%%bigquery
select sum(safe_cast(ell as int64)) as ell, district, year, grade
from school_enrollments.ca
where grade != 'k_12_total'
group by district, year, grade
having ell > 1000
order by district, year, grade
```

### 8. Find the number of students with disabilities in each district in Connecticut, listing only those with more than 50 disabilities and only showing the top 30 districts.

```python
%%bigquery
select sum(safe_cast(disability as int64)) as disability, district
from school_enrollments.ct
where year = 2021 and admin_level != 'district'
group by district
having disability > 50
order by disability DESC
limit 30
```

### 9. Find the number of male, female, and total students for California and Texas per school, filtered for schools with >5000 students

```python
%%bigquery
select state, school, year, sum(safe_cast(male as int64)) as male, sum(safe_cast(female as int64)) as female, sum(safe_cast(total as int64)) as total
from school_enrollments.ca
where admin_level = 'school'
group by state, school, year
having total > 5000
UNION ALL
select state, school, year, sum(safe_cast(male as int64)) as male, sum(safe_cast(female as int64)) as female, sum(safe_cast(total as int64)) as total
from school_enrollments.tx
where admin_level = 'school'
group by state, school, year
having total > 5000
order by total desc
```
